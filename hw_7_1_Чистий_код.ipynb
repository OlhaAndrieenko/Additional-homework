````python
from math import *
                                                #  додано 1 порожній рядок
def calc_area(r):                               # перейменовано функцію
    return pi * r**2                              # блок коду перенесено на нову строку, додано відступ, пробіли між оператором
                                                # між визначенням функцій додано 2 порожніх рядка

def calc_сircumference(r):                       # перейменовано функцію
    return 2 * pi * r                            # блок коду перенесено на нову строку, додано відступ, пробіли між оператором

radius = 5                                       #додано пробіли при визначенні змінної
print("Area:", calc_area(radius), "Circumference:", calc_сircumference(radius))    # додано пробіли між значеннями
````

І «product_name», і «product_category» є рядками, що представляють численні, але обмежені унікальні назви продуктів і категорії відповідно.

Напишіть код для ефективного перетворення стовпців «product_name» і «product_category» на категорійний тип для ефективного використання пам’яті.
Спробуйте вирішити це завдання, мінімізуючи використання пам'яті та зберігаючи читабельність коду. Крім того, виміряйте використання пам’яті до та після перетворення, щоб продемонструвати підвищення ефективності.

````python
import pandas as pd
import numpy as np

np.random.seed(0)

products = ['apple', 'banana', 'cherry', 'dates', 'elderberry']
categories = ['fruit', 'fruit', 'fruit', 'fruit', 'fruit']

N_ROWS = 10**6
df = pd.DataFrame({
    'product_name': np.random.choice(products, N_ROWS),
    'product_category': np.random.choice(categories, N_ROWS),
    'product_price': np.random.uniform(1, 10, N_ROWS),
    'transaction_date': pd.date_range(start='1/1/2000', periods=N_ROWS, freq='h')
})

initial_memory = df.memory_usage(index=False).sum()
print(f"Початкове використання пам'яті: {initial_memory}")
````

````python
df['product_name'] = df['product_name'].astype('category')
df['product_category'] = df['product_category'].astype('category')

optimized_memory = df.memory_usage(index=False).sum()
print(f"Використання пам'яті після оптимізації: {optimized_memory}")
````

````python
def calculate_variance(nl):
    """Обчислює дисперсію для списку чисел."""
    mean = sum(nl) / len(nl)
    mean_squares = sum(n**2 for n in nl) / len(nl)
    variance = mean_squares - mean ** 2  # Дотримуємося PEP 8 (пробіли навколо **)
    return variance


# Викликаємо функцію
input_list = [75, 80, 85, 90, 95]
variance = calculate_variance(input_list)
print("Дисперсія:", variance)
````
